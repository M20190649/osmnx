### OSMからoverpass APIを経由して、特定のNodeを取得して地図上に表示する

六本木周辺のコンビニとバス停を抽出し、それぞれ地図にプロットする

```python
import folium
import overpy


# 六本木周辺を指定 
bbox = 	'35.65604,139.71775,35.67584,139.75641'
bounds = [[35.65604,139.75641],[35.67584,139.71775]]
api = overpy.Overpass()

# 周辺のコンビニ
poi = [(35.664333, 139.732863), 
           (35.661785, 139.735662),
           (35.663956, 139.738678),
           (35.671530, 139.741861),
           (35.672438, 139.742671),
           (35.671773, 139.735621),
           (35.674275, 139.737802)]
```

```python
# コンビニを抽出
conv_query = 'node["shop"="convenience"]({bbox});out body;'.format(bbox = bbox)

# 実行
c = api.query(conv_query)

```

```python
#コンビニ

conv_map = folium.Map(tiles='cartodbpositron')
conv_map.fit_bounds(bounds)

for i in c.nodes:
    if 'name' in i.tags.keys():
        name = i.tags['name']
    elif 'brand' in i.tags.keys():
        name = i.tags['brand']
    if 'branch' in i.tags.keys():
        name += i.tags['branch']
    folium.Marker(
        [i.lat, i.lon], 
        popup=folium.Popup("{}({})".format(name, i.id), parse_html=True),
        icon=folium.Icon(color='green')
    ).add_to(conv_map)

conv_map
```
