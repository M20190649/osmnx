### OSMからoverpass APIを経由して、特定のNodeを取得して地図上に表示する

六本木周辺のコンビニとバス停を抽出し、それぞれ地図にプロットする

```python
import folium
import overpy


# 六本木周辺を指定 
bbox = 	'35.65604,139.71775,35.67584,139.75641'
bounds = [[35.65604,139.75641],[35.67584,139.71775]]
api = overpy.Overpass()
```

```python
# コンビニを抽出
conv_query = 'node["shop"="convenience"]({bbox});out body;'.format(bbox = bbox)

# バス停を抽出
bus_query = '(node["highway"="bus_stop"]({bbox}); \
                        way["highway"="bus_stop"]({bbox}); \
                        relation["highway"="bus_stop"]({bbox}); );\
                        out body;'.format(bbox = bbox)

# 実行
c = api.query(conv_query)
b = api.query(bus_query)
```

```python
#コンビニ

conv_map = folium.Map(tiles='cartodbpositron')
conv_map.fit_bounds(bounds)

for i in c.nodes:
    if 'name' in i.tags.keys():
        name = i.tags['name']
    elif 'brand' in i.tags.keys():
        name = i.tags['brand']
    if 'branch' in i.tags.keys():
        name += i.tags['branch']
    folium.Marker(
        [i.lat, i.lon], 
        popup=folium.Popup("{}".format(name), parse_html=True),
        icon=folium.Icon(color='green')
    ).add_to(conv_map)

conv_map
```

```python
## バス停

bus_map = folium.Map(tiles='cartodbpositron')
bus_map.fit_bounds(bounds)

for i in b.nodes:
    if 'name' in i.tags.keys():
        name = i.tags['name']
    elif 'brand' in i.tags.keys():
        name = i.tags['brand']
    if 'branch' in i.tags.keys():
        name += i.tags['branch']
    folium.Marker(
        [i.lat, i.lon], 
        popup=folium.Popup("{}".format(name), parse_html=True),
        icon=folium.Icon(color='blue',icon='info-sign')
    ).add_to(bus_map)

bus_map
```
